### Predicting Undergrad Tuition and Fees at Private Colleges/Universities

The goal of this project is to identify which features of private non-profit institutions in the U.S. that offer Bachelor's degrees can best predict undergraduate tuition and fees at these institutions.

I began the analysis by exploring the one-to-one relationship between undergraduate tuition and fees at over 1,000 schools and various features of these schools. I discovered that the strongest was with the average amount of grant or scholarship aid received by undergraduate students. Therefore, I created a simple linear regression model to predict undergraduate tuition and fees based on average amount of undergraduate grant or scholarship aid (see plot below).

The plot illustrates that while the average amount of undergraduate grant or scholarship aid appears to do a fairly good job at predicting undergraduate tuition and fees, the distances between some of the actual tuition and fees (represented by the black dots) and the predicted amounts (represented by the green line) indicate that the model can be improved upon in order to more accurately predict tuition and fees.
